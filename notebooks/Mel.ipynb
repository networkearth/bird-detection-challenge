```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = 'Haunting_song_of_humpback_whales.wav'
y, sr = librosa.load(filename)

whale_song, _ = librosa.effects.trim(y)
librosa.display.waveshow(whale_song, sr=sr)
```

```python
n_fft = 2048
D = np.abs(librosa.stft(whale_song[:n_fft], n_fft=n_fft, hop_length=n_fft+1))
plt.plot(D);
```

```python
hop_length = 512
D = np.abs(librosa.stft(whale_song, n_fft=n_fft, hop_length=hop_length))
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='linear');
plt.colorbar();
```

```python
DB = librosa.amplitude_to_db(D, ref=np.max)
librosa.display.specshow(DB, sr=sr, hop_length=hop_length, x_axis='time', y_axis='linear')
plt.colorbar(format='%+2.0f dB');
```

```python
n_mels = 40
mel = librosa.filters.mel(sr=sr, n_fft=n_fft, n_mels=n_mels)
```

```python
S = librosa.feature.melspectrogram(whale_song, sr=sr, n_fft=n_fft, hop_length=hop_length, n_mels=n_mels, power=1)
S_DB = librosa.amplitude_to_db(S, ref=np.max)
librosa.display.specshow(S_DB, sr=sr, hop_length=hop_length, x_axis='time', y_axis='mel');
plt.colorbar(format='%+2.0f dB');
```
